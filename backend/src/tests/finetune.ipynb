### Parse PDF Files 
Extract text from PDF files

```python
import os
import sys
from pathlib import Path

# add the parent directory to the system path
current_dir = Path.cwd()
parent_dir = str(current_dir.parent) 
sys.path.append(parent_dir)

from scripts.pdf_convert import convert_pdf
```

```python
def load_and_convert_data():
    """ 
    Loads the data from the dataset folder and converts all pdf files to text.
    """
    # read the dataset folder
    dat_dir = "../../../data/raw"
    files = os.listdir(dat_dir)  # read all files in the dataset folder
    
    # converts all pdf files to text and stores them in an array
    texts = [
        convert_pdf(Path(dat_dir, file).read_bytes()) 
        for file in files 
        if file.endswith('.pdf')
    ]

    print("text in pdf: ", texts)

    return texts

# parameter for fine-tuning
texts = load_and_convert_data()
```

### Import and fine-tune the model

```python
# Import dependencies
import openai
import os
from dotenv import load_dotenv

load_dotenv()

client = openai.OpenAI(api_key = os.getenv("OPENAI_API_KEY"))
```

```python
def fine_tune_model(texts):
    """
    Fine-tunes the model with the given texts.
    """
    
    response = client.fine_tuning.jobs.create(
        training_file = texts,
        model = "gpt-3.5-turbo",
        hyperparameters = {
            "n_epochs": 3 
        }
    )

    return response

fine_tune_model(texts)
```
